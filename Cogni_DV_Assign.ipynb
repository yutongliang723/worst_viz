```python
from PIL import Image
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 40))  
ax = fig.add_subplot(111, projection='3d')

img = data.rocket()
img = Image.open('background_image.png')
ax.imshow(img, extent=[-0.1, 0.1, -0.1, 0.1], origin = 'upper', alpha = 0.45)

bar_width = 10
for i in range(len(mtcars['mpg'])):
    x = mtcars['mpg'][i]
    y = mtcars['disp'][i]
    z = 0  
    height = mtcars['hp'][i]
    ax.bar3d(x, y, z, bar_width, bar_width, height, color='m', alpha=0.5)

ax.set_xlabel('MPG', fontsize=5)  
ax.set_ylabel('Displacement', fontsize=5)
ax.set_zlabel('Horsepower', fontsize=5)
ax.set_title('Car Attributes', fontsize=5)    
ax.view_init(elev=10, azim=50)
ax.set_facecolor('yellow')
# n_grids = 20
ax.xaxis._axinfo['grid'].update(color='red', linestyle='--', linewidth=0.5)
ax.yaxis._axinfo['grid'].update(color='red', linestyle='--', linewidth=0.5)
ax.zaxis._axinfo['grid'].update(color='red', linestyle='--', linewidth=0.5)
n_grids = 20
for i in range(0, len(mtcars)):
    ax.text(mtcars['mpg'][i], mtcars['disp'][i], mtcars['hp'][i], f'{i}', color='blue', fontsize=15)

ax.xaxis.set_major_locator(plt.MaxNLocator(n_grids))
ax.yaxis.set_major_locator(plt.MaxNLocator(n_grids))
ax.zaxis.set_major_locator(plt.MaxNLocator(n_grids))


plt.show()
```
